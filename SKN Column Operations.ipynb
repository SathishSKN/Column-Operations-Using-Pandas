### This dataset is downloaded from the Kaggle dataset website and used it to do some basic data exploration using pandas. This dataset contains the actions contained in a market area

#### To perform the operations we need to import a library, so here pandas library is imported locally as pd

```python
Happy = pd.read_csv("D://MBA//SEM 2//Python//World_happiness_2019.csv")
Happy
```

#### Use pd.read_csv function to read the dataset in to Jupyter Notebook

```python
Happy['Happiness'] = Happy['Freedom to make life choices'] + Happy['Generosity']
Happy
```

#### A new variable called "Happiness" is added to the dataset with help of using the existing variables which are "Freedom to make life choices" & "Generosity".

```python
Happy.rename(columns = {'Happiness' : 'Happier'})
```

#### A single variable Happiness is renamed as Happier using rename function.

```python
Happy.rename(columns = {'Score' : 'Score_Value','Freedom to make life choices' : 'Freedom','Generosity' : 'Generous'})
```

#### Multiple variable name is renamed here using rename function like earlier

```python
Happy.drop(['Happiness'], axis=1)
```

#### A variable called Happiness is removed using the drop function

```python
Happy.drop(['Social support','Freedom to make life choices','Happiness'], axis=1)
```

#### Multiple variable is removed using the drop function

```python
Happy.columns = [col.upper() for col in Happy.columns]
Happy
```

#### Uppercase of columns name

```python
Happy.columns = [col.lower() for col in Happy.columns]
Happy
```

#### Lowercase of coulmns name

```python
Happy.sort_values(by = 'score', ascending = True)
```

#### Sorting the entire dataset ascendingly with respect to variable "Score". 

```python
Happy.sort_values(by = ['score','generosity'], ascending =[True,False])
```

#### Sorting multiple variables using sort function

```python
Happy.drop(['social support','generosity','happiness'], axis=1,inplace = True)
Happy
```

```python
Happy.groupby('score').count()
```
